```python
import numpy as np
import matplotlib.pyplot as plt

data = [1,0,1,1,0]
bit_duration = 1
fs = 1000
t=np.arange(0,len(data)*bit_duration,1/fs)
f1=5
f2=10
A1=0
A2=1

def ask(data,t):
    signal = np.zeros(len(t))
    for i ,bit in enumerate(data):
        amp = A2 if bit else A1
        signal[i*fs:(i+1)*fs] = amp*np.sin(2*np.pi*amp*t[i*fs:(i+1)*fs])

    return signal

def fsk(data,t):
    signal = np.zeros(len(t))
    for i ,bit in enumerate(data):
        freq = f2 if bit else f1
        signal[i*fs:(i+1)*fs] = np.sin(2*np.pi*freq*t[i*fs:(i+1)*fs])

    return signal

def psk(data,t):
    signal = np.zeros(len(t))
    for i ,bit in enumerate(data):
        phase = 0 if bit else np.pi
        signal[i*fs:(i+1)*fs] = np.sin(2*np.pi*f1*t[i*fs:(i+1)*fs]+phase)

    return signal

plt.figure(figsize=(12,8))
plt.subplot(3,1,1)
plt.title("ASK modulation")
plt.plot(t,ask(data,t))
plt.grid(True)
plt.figure(figsize=(12,8))
plt.subplot(3,1,1)
plt.title("FSK modulation")
plt.plot(t,fsk(data,t))
plt.grid(True)
plt.figure(figsize=(12,8))
plt.subplot(3,1,1)
plt.title("PSK modulation")
plt.plot(t,psk(data,t))
plt.grid(True)


```
