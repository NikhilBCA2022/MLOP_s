```python
##inheritance

class car:
    def __init__(self,brand,name,windows):
           self.windows=windows
           self.brand=brand
           self.name=name

    
    def drive(self):
          print(f"the person will drive {self.name}")

    
car1=car("BWM","BWM","POlO")

car1.drive()
        
        
```

```python
class Tesla(car):
    def __init__(self, brand, name, windows,is_self):
        super().__init__(windows,brand,name)
        self.is_self=is_self

    def selfdrive(self):
        print(f"this is {self.is_self}")

tesla1=Tesla("BME","4","MOTO",True)

tesla1.selfdrive()
        
```

```python
##multiple inheritance

class animal:
    def __init__(self,name):
        self.name=name

class dog:
    def __init__(self,owner):
        self.owner=owner

class pet(dog,animal):
    def __init__(self, name,owner):
        animal.__init__(self,name)
        dog.__init__(self,owner)

    def speak(self):
        return f"the dog name is {self.name} and owner is {self.owner}"
    

dog1=pet("buddy","nikhil")

dog1.speak()

dog1.owner




```
