```python
##inheritance

class car:
    def __init__(self,brand,name,windows):
           self.windows=windows
           self.brand=brand
           self.name=name

    
    def drive(self):
          print(f"the person will drive {self.name}")

    
car1=car("BWM","BWM","POlO")

car1.drive()
        
        
```

```python
class Tesla(car):
    def __init__(self, brand, name, windows,is_self):
        super().__init__(windows,brand,name)
        self.is_self=is_self

    def selfdrive(self):
        print(f"this is {self.is_self}")

tesla1=Tesla("BME","4","MOTO",True)

tesla1.selfdrive()
        
```
