```python
##check whether the year is leap year or not

year = int(input("enter the year : "))

if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year,"it is leap year")
        else:
            print(year,"it is not leap year")
    else:
        print(year,"it is leap year")
else:
    print(year,"is leap year")


```
